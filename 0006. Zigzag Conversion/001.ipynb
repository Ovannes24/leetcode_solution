```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# s = "PAYPALISHIRING"
s = np.arange(100)
l = list(range(len(s)))
n = 7
```

```python
am = [
    l[0::2*n-2],
    sum([
        sorted(l[i+1::2*n-2] + l[(i+1)+2*n-2-2*(i+1)::2*n-2])
        for i in range(n-2)
    ], []),
    l[n-1::2*n-2]
    
]
```

```python
ns = [s[i] for i in sum(am, [])]
ns
```

```python
plt.plot(s, ns, '.-')
plt.grid()
```
